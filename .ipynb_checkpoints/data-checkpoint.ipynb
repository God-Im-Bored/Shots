```python
from nba_api.stats.endpoints import shotchartdetail
import pandas as pd
import json

response = shotchartdetail.ShotChartDetail(
        team_id=0,
        player_id=202695,
        season_nullable='2020-21',
        season_type_all_star='Regular Season'
)
data = json.loads(response.get_json())

results = data['resultSets'][0]
headers = results['headers']
rows = results['rowSet']
df = pd.DataFrame(rows)
df.columns = headers
df.to_csv(r'/home/aram/Downloads/kawhi_21.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('white')
sns.set_color_codes()
plt.figure(figsize=(12,11))
plt.scatter(df.LOC_X, df.LOC_Y)
plt.show()
```

```python
from matplotlib.patches import Circle, Rectangle, Arc

def court(ax=None, color='black', lw=2, outer_lines=False):
    if ax is None:
        ax = plt.gca()
    hoop = Circle((0, 0), radius=7.5, linewidth=lw, color=color, fill=False)
    backboard = Rectangle((-30, -7.5), 60, -1, linewidth=lw, color=color)
    outer_box = Rectangle((-80, -47.5), 160, 190, linewidth=lw, color=color,
                          fill=False)
    inner_box = Rectangle((-60, -47.5), 120, 190, linewidth=lw, color=color,
                          fill=False)
    
    top_free_throw = Arc((0, 142.5), 120, 120, theta1=0, theta2=180,
                         linewidth=lw, color=color, fill=False)
    
    bottom_free_throw = Arc((0, 142.5), 120, 120, theta1=180, theta2=0,
                            linewidth=lw, color=color, linestyle='dashed')

    restricted = Arc((0, 0), 80, 80, theta1=0, theta2=180, linewidth=lw,
                     color=color)

    corner_three_a = Rectangle((-220, -47.5), 0, 140, linewidth=lw,
                               color=color)
    corner_three_b = Rectangle((220, -47.5), 0, 140, linewidth=lw, color=color)
   
    three_arc = Arc((0, 0), 475, 475, theta1=22, theta2=158, linewidth=lw,
                    color=color)

    center_outer_arc = Arc((0, 422.5), 120, 120, theta1=180, theta2=0,
                           linewidth=lw, color=color)
    center_inner_arc = Arc((0, 422.5), 40, 40, theta1=180, theta2=0,
                           linewidth=lw, color=color)

    court_elements = [hoop, backboard, outer_box, inner_box, top_free_throw,
                      bottom_free_throw, restricted, corner_three_a,
                      corner_three_b, three_arc, center_outer_arc,
                      center_inner_arc]

    if outer_lines:
        
        outer_lines = Rectangle((-250, -47.5), 500, 470, linewidth=lw,
                                color=color, fill=False)
        court_elements.append(outer_lines)

    for element in court_elements:
        ax.add_patch(element)

    return ax
```

```python
plt.figure(figsize=(12,11))
court(outer_lines=True)
plt.xlim(-300,300)
plt.ylim(-100,500)
plt.show()
```

```python
plt.figure(figsize=(12,11))
plt.scatter(df.LOC_X, df.LOC_Y)
court(outer_lines=True)
plt.xlim(300, -300)
plt.show()
```

```python
plt.figure(figsize=(12,11))
plt.scatter(df.LOC_X, df.LOC_Y)
court()
plt.xlim(-250,250)
plt.ylim(422.5, -47.5)
plt.show()
```

```python
import urllib.request

pic = urllib.request.urlretrieve('https://cdn.nba.com/headshots/nba/latest/1040x760/202695.png')

kawhi_pic = plt.imread(pic[0])

plt.imshow(kawhi_pic)
plt.show()
```

```python
from matplotlib.offsetbox import  OffsetImage
cmap=plt.cm.Blues
shot_chart = sns.jointplot(df.LOC_X, df.LOC_Y, kind='scatter', space=0, color=cmap(.8), cmap=cmap)
shot_chart.fig.set_size_inches(12,11)

ax = shot_chart.ax_joint
court(ax)

ax.set_xlim(-250,250)
ax.set_ylim(422.5, -47.5)


ax.set_xlabel('')
ax.set_ylabel('')
ax.tick_params(labelbottom='off', labelleft='off')

ax.set_title('Kawhi Leonard FGA \n2020-21 Reg. Season', y=1.2, fontsize=18)

ax.text(-250, 465, 'Data source: stats.nba.com' '\nAuthor: Aram Martin', fontsize=12)

img = OffsetImage(kawhi_pic, zoom=0.2)

img.set_offset((625, 650))

ax.add_artist(img)

plt.show()


```
